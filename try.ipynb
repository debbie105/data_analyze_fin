```
import requests
from bs4 import BeautifulSoup
url="https://www.ptt.cc/bbs/marvel/index2434.html"


def get_marvel(url):
     r = requests.get(url)#將網頁資料GET下來
     sp = BeautifulSoup(r.text,"html.parser")#將網頁資料以html.parser
     rs = sp.select('.r-ent')
     for item in rs:
          print ('日期:',item.select('.date')[0].text)
          print ('推文數:',item.select('.nrec')[0].text)
          print ('作者:',item.select('.author')[0].text)
          print ('標題:',item.select('.title')[0].text)
          print ('url:',item.select('a')[0])
          print ('==============================================')


for page in range(0,4):
    r = requests.get(url)
    sp = BeautifulSoup(r.text,"html.parser")
    btn = sp.select('div.btn-group > a')
    up_page_href = btn[3]['href']
    next_page_url = 'https://www.ptt.cc' + up_page_href
    url = next_page_url
    get_marvel(url = url)


```
